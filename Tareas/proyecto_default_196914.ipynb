# Responde las siguientes preguntas generales (sklearn)

1. Las metricas (todas, algunas?) de sub, over, ponderadas y sin sampleo son comparables entre si o no? (Cuales son comparables entre si y bajo que criterios?)

Las métricas entre diferentes estrategias de muestreo y ponderación pueden ser comparables hasta cierto punto, pero hay ciertas consideraciones importantes. Algunas métricas como la precisión, la sensibilidad (recall), y el F1-score son sensibles al desbalanceo de clases y pueden ofrecer una visión comparativa valiosa entre diferentes estrategias. Por otro lado, la AUC-ROC es una métrica que es generalmente robusta al desbalanceo de clases, proporcionando una medida comparativa fiable del rendimiento del modelo entre diferentes estrategias.

Si ya tratamos los datos, las métricas que son sensibles al desbalanceo se ajustarán proporcionalmente a los tratamientos, por ello, para poder comparar estos aproaches debemos usar una que sensible.

- **Métricas Sensibles al Desbalanceo**:
  * *Precisión*: En un escenario desbalanceado, si un modelo predice la clase mayoritaria todo el tiempo, la precisión puede ser alta, lo que no refleja la verdadera capacidad del modelo para identificar la clase minoritaria.
  * *Sensibilidad (Recall)*: Aunque esta métrica se centra en la clase positiva, puede ser influenciada por un desbalanceo severo de clases.
  * *F1-Score*: El F1-Score es la media armónica de precisión y sensibilidad. Aunque combina ambas métricas, aún puede ser afectado por el desbalanceo de clases, especialmente si el modelo tiene un rendimiento pobre en la clase minoritaria.
- **Métricas Robustas al Desbalanceo**:
  * *AUC-ROC*: Es una métrica que evalúa la capacidad del modelo de distinguir entre clases. Esta métrica es considerada robusta frente al desbalanceo de clases porque evalúa el rendimiento del modelo en todos los umbrales de clasificación, en lugar de un umbral específico. Esto la hace menos susceptible a la influencia del desbalanceo de clases ya que integra el rendimiento del modelo a lo largo de todos los posibles umbrales de clasificación.

2. ¿Debes aplicar sub, over sampleo y poderacion al conjunto de test para poder comparar las metricas que se crearon durante el entrenamiento con las de test?

* *Subsamplig*: DUDA No es adecuado aplicar submuestreo en el conjunto de test. El conjunto de test debe representar la distribución real de las clases en la población para proporcionar una evaluación imparcial y realista del rendimiento del modelo. El submuestreo en el conjunto de test alteraría la distribución natural de las clases y podría resultar en una evaluación engañosa del rendimiento del modelo. La idea del conjunto de test es evaluar cómo el modelo se desempeñaría en datos no vistos que reflejan la realidad.
* *Oversampling*: El sobremuestreo implicaría replicar las instancias de la clase minoritaria en el conjunto de test, lo cual tampoco reflejaría la distribución real de las clases. Además, podría inflar artificialmente el rendimiento del modelo en la clase minoritaria, proporcionando una visión inexacta del rendimiento del modelo.
* *Ponderación*: Aplicar ponderación en el conjunto de test no sería apropiado ni estándar en la práctica de machine learning. La ponderación está destinada a ayudar al modelo a aprender mejor de la clase minoritaria durante el entrenamiento. El conjunto de test debe permanecer sin modificar para proporcionar una evaluación precisa del rendimiento del modelo.

El propósito del conjunto de datos de prueba es evaluar el rendimiento del modelo en condiciones que reflejen lo más fielmente posible la realidad del problema que se está tratando de resolver. En el caso de la validación cruzada se menciona la posibilidad de subsamplear y validar cruzadamente los resultados de los modelos, pero se refiere a hacerlo durante el proceso de entrenamiento y validación.

3. Que metricas (incluyendo la matriz confusion) te parecen las mejores para elegir resultados y decidir qué modelo entrara a produccion? (y por que?)

* *AUC-ROC (Área bajo la Curva - Característica Operativa del Receptor)*: Esta métrica es valiosa ya que proporciona una visión global del rendimiento del modelo independientemente del umbral de clasificación utilizado.
* *F1-Score*: El F1-Score es una métrica que combina tanto la precisión como la sensibilidad (recall) en una sola métrica, proporcionando un balance entre los falsos positivos y los falsos negativos.
* *Matriz de Confusión*: Proporciona un desglose detallado de los verdaderos positivos, verdaderos negativos, falsos positivos y falsos negativos, lo cual es crucial para entender el desempeño del modelo en un nivel más granular.

```python
def _check_type(value):
    '''
    Infiere si la variable es categorica o numerica
    apartir del tipo
    '''
    if value==object:
        return 'categorica'
    elif value==bool:
        return 'categorica'
    elif value=='int64':
        return 'numerica'
    elif value=='float64':
        return 'numerica'
```

## Elección de Métodos de Ponderación
**La ponderación de clases y la ponderación de muestras** fueron elegidas como métodos para abordar el desbalanceo de clases en nuestro conjunto de datos. La ponderación de clases se realiza proporcionando un peso diferente a cada clase durante el entrenamiento, lo que ayuda a garantizar que la clase minoritaria no se ignore. La ponderación de muestras, por otro lado, proporciona un peso diferente a cada muestra individual, lo que permite un control más granular sobre cómo se maneja el desbalanceo.

Los pesos de la muestra se calcularon usando `sklearn.utils.class_weight.compute_sample_weight` con la opción `balanced`, que automáticamente ajusta los pesos inversamente proporcional al número de muestras de cada clase.

## ¿Aplicación en el Conjunto de Datos de Prueba?
Es importante notar que estos métodos de ponderación y reequilibrio solo deben aplicarse al conjunto de datos de entrenamiento. El conjunto de datos de prueba debe permanecer sin modificar para proporcionar una evaluación imparcial del rendimiento del modelo en datos no vistos. Por lo tanto, no se aplican pesos ni reequilibrio en el conjunto de datos de prueba en este código.




```python
from sklearn.utils.class_weight import compute_sample_weight
#from sklearn.utils.class_weight import compute_class_weight


# Calcular los pesos de las muestras
sample_weights = compute_sample_weight(class_weight='balanced', y=df_train['y'])

# Calcular los pesos de las clases manualmente
weights = compute_sample_weight(class_weight='balanced', y=df_train['y'])
class_weight_dict = {0: np.unique(weights)[0], 1: np.unique(weights)[1]}
```

### Código modificado

**Ponderaciones de Clases (class_weight)**: Se utiliza para modelos que permiten el parámetro class_weight (como LogisticRegression y RandomForestClassifier). Este parámetro ajusta la función de pérdida durante el entrenamiento para dar más importancia a las clases minoritarias.

**Ponderaciones de Muestras (sample_weight)**: Se utiliza para modelos que permiten el parámetro sample_weight en su método fit. Este parámetro ajusta la función de pérdida durante el entrenamiento para dar más importancia a las muestras específicas. En este caso, se da más peso a las muestras de la clase minoritaria.

La función model_selection ha sido modificada para verificar si el modelo actual acepta sample_weight como un parámetro en su método fit. Si es así, las ponderaciones de las muestras se pasan durante el entrenamiento. Si no, el modelo se entrena normalmente sin las ponderaciones de las muestras. Esto permite utilizar una mezcla de modelos que admitan o no ponderaciones de muestras, maximizando así la utilidad de las ponderaciones calculadas sin restringir la selección de modelos.

*Nota:*
 - Linear Discriminant Analysis (LinearDiscriminantAnalysis): No permite el uso de sample_weight ni class_weight. Este modelo se basa en la suposición de que las clases tienen la misma covarianza y maximiza la distancia entre las medias de las clases, sin proporcionar una forma directa de ponderar las clases o las muestras durante el entrenamiento.

- Neural Networks (MLPClassifier): No permite el uso de class_weight, pero sí permite el uso de sample_weight en el método fit. Por lo tanto, puedes utilizar ponderaciones a nivel de muestra para manejar el desbalanceo de clases durante el entrenamiento de redes neuronales en scikit-learn.



```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold, cross_validate
import pandas as pd
import seaborn as sns

def model_selection_v2(df_train, num_folds=3, seed=7364, scoring=["accuracy","balanced_accuracy","roc_auc", "f1"]):
    X_train = df_train.drop(columns=['y'])
    Y_train = df_train['y']

    # Normalización
    from sklearn.preprocessing import MinMaxScaler
    mm_scaler = MinMaxScaler()
    X_train = mm_scaler.fit_transform(X_train)

    # Modelos
    models = []
    models.append(('LR', LogisticRegression(max_iter=500, class_weight=class_weight_dict)))
    models.append(('RF', RandomForestClassifier(class_weight=class_weight_dict)))
    models.append(('KNN', KNeighborsClassifier()))
    models.append(('GBM', GradientBoostingClassifier()))

    # Validación cruzada y resultados
    results = []
    for name, model in models:
        print(f"Model: {name}")
        kfold = KFold(n_splits=num_folds, random_state=seed, shuffle=True)

        # Si el modelo soporta 'sample_weight', usarlo durante el entrenamiento
        if "sample_weight" in model.fit.__code__.co_varnames:
            cv_results = cross_validate(model, X_train, Y_train, cv=kfold,
                                        scoring=scoring, n_jobs=-1,
                                        fit_params={'sample_weight': sample_weights})
        else:
            cv_results = cross_validate(model, X_train, Y_train, cv=kfold,
                                        scoring=scoring, n_jobs=-1)

        cv_results = pd.DataFrame(cv_results)
        cv_results.loc[:,'model_name'] = name
        results.append(cv_results)

    # Visualización de resultados
    results = pd.concat(results, ignore_index=True)
    print(results.groupby(['model_name'], as_index=False).describe())
    results_plot = [c for c in results.columns if 'time' not in c]
    results_plot = results.loc[:,results_plot]
    results_plot = pd.melt(results_plot, id_vars='model_name', var_name="metric")
    g = sns.FacetGrid(results_plot, row="metric", height=3, aspect=1, sharex=False)
    g.map(sns.boxplot, 'model_name','value')
    g.fig.set_size_inches(13, 20)

    return results
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import cross_validate, KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.utils.class_weight import compute_sample_weight

def model_selection_v2(df_train, num_folds=3, seed=7364, scoring=["accuracy", "balanced_accuracy", "roc_auc", "f1"]):
    X_train = df_train.drop(columns=['y'])
    Y_train = df_train['y']

    # Normalización
    mm_scaler = MinMaxScaler()
    X_train = mm_scaler.fit_transform(X_train)

    # Modelos
    models = [
        ('LR', LogisticRegression(max_iter=500)),
        ('LDA', LinearDiscriminantAnalysis()),
        ('KNN', KNeighborsClassifier()),
        ('NN', MLPClassifier(max_iter=500, solver='sgd', early_stopping=True)),
        ('GBM', GradientBoostingClassifier()),
        ('RF', RandomForestClassifier())
    ]

    # Calculando los pesos de las muestras
    sample_weights = compute_sample_weight(class_weight='balanced', y=Y_train)

    # Validación cruzada y resultados
    results = []
    for name, model in models:
        class_weights = [None]
        if "class_weight" in model.get_params():
            class_weights.append('balanced')

        for class_weight in class_weights:
            print(f"Model: {name}, Class Weight: {class_weight}")
            if class_weight is not None:
                model.set_params(class_weight=class_weight)

            kfold = KFold(n_splits=num_folds, random_state=seed, shuffle=True)

            fit_params = {}
            if "sample_weight" in model.fit.__code__.co_varnames:
                fit_params['sample_weight'] = sample_weights

            cv_results = cross_validate(model, X_train, Y_train, cv=kfold,
                                        scoring=scoring, n_jobs=-1, return_train_score=True,
                                        fit_params=fit_params)

            cv_results = pd.DataFrame(cv_results)
            cv_results.loc[:,'model_name'] = f'{name}_{"balanced" if class_weight else "unbalanced"}'
            results.append(cv_results)

    # Visualización de resultados
    results = pd.concat(results, ignore_index=True)

    for metric in scoring:
        plt.figure(figsize=(13, 6))

        # Plotting test metrics
        test_data = results[['model_name', f'test_{metric}']]
        test_data = test_data.rename(columns={f'test_{metric}': metric})
        test_data['phase'] = 'Test'

        # Plotting train metrics
        train_data = results[['model_name', f'train_{metric}']]
        train_data = train_data.rename(columns={f'train_{metric}': metric})
        train_data['phase'] = 'Train'

        # Concatenating the data
        plot_data = pd.concat([test_data, train_data], axis=0)

        sns.boxplot(data=plot_data, x='model_name', y=metric, hue='phase')
        plt.title(f'{metric} Comparison')
        plt.grid(True)
        plt.show()

    return results
```
